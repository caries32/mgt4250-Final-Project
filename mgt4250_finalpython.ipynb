```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('/Users/cameronaries/Downloads/washington_state_evs.csv')

# Extracting year from the 'Date' column
df['Year'] = pd.to_datetime(df['Date']).dt.year

# Grouping by year and summing the total number of electric vehicles and total vehicles
total_by_year = df.groupby('Year').agg({'Electric Vehicle (EV) Total': 'sum', 'Total Vehicles': 'sum'}).reset_index()

# Calculating the correct percentage of electric vehicles
total_by_year['Percent Electric Vehicles'] = (total_by_year['Electric Vehicle (EV) Total'] / total_by_year['Total Vehicles']) * 100

# Creating a bar chart for the percentage of electric vehicles by year
plt.figure(figsize=(10, 6))
plt.bar(total_by_year['Year'], total_by_year['Percent Electric Vehicles'], color='deepskyblue')
plt.title('Percentage of Electric Vehicles by Year in Washington State')
plt.xlabel('Year')
plt.ylabel('Percentage of Electric Vehicles')
plt.xticks(total_by_year['Year'])
plt.grid(axis='y')

# Display the bar chart
plt.show()
```
